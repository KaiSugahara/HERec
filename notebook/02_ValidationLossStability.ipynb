# Validation Loss Stability

```python
import sys
sys.path.append("..")

import numpy as np
import matplotlib.pyplot as plt

from herec.utils import *
```

## On Explicit Feedback

```python
model_names = ["MF", "HE_MF"]
dataset_names = ["ML100K", "ML1M", "Ciao_PART", "Ciao"]
```

```python
for scale_type in ["linearScale", "logScale"]:

    fig, ax = plt.subplots(nrows=4, ncols=1, sharex=True, sharey=True, figsize=(7, 7))
    
    for ax_i, dataset_name in enumerate(dataset_names):
    
        for pos_i, model_name in enumerate(model_names):
    
            # Get (best) valid. loss
            loader = resultLoader(f"HeRec-{model_name}-{dataset_name}")
            valid_losses = loader.result_df.get_column("metrics.BEST_VALID_LOSS").to_numpy()
            valid_scores = np.sqrt(valid_losses)
            print("Trial #:", len(valid_scores))
            
            # BoxPlot
            ax[ax_i].boxplot( valid_scores, vert=False, widths=0.6, whis=(0, 100), labels=[model_name.replace("_", "-")], positions = [-pos_i] )
    
            # Scatter
            ax[ax_i].scatter( valid_scores, np.ones_like(valid_losses)*(-pos_i), s=20, c="red" )

            # Adjust Layout
            if scale_type == "logScale": ax[ax_i].set_xscale( 'log' )
            ax[ax_i].set_xlabel( 'RMSE on Valid. Subset' )
            ax[ax_i].set_ylabel( 'Model' )
            ax[ax_i].set_title( dataset_name.replace("_", "-") )
            ax[ax_i].xaxis.set_tick_params(labelbottom=True)

    plt.suptitle( scale_type )
    plt.tight_layout()
    plt.savefig( f"01_ValidationLossStability/explicitDataset_boxPlot_{scale_type}.pdf" )
    plt.show()
```

## On Implicit Feedback

```python
model_names = ["MF_BPR", "HE_MF_BPR"]
dataset_names = ["Twitch100K", "DIGINETICA", "ML100K_IMPLICIT", "ML1M_IMPLICIT"]
metrics = ["Recall_10", "Recall_30", "nDCG_10", "nDCG_30"]
```

```python
for metric in metrics:
    for scale_type in ["linearScale", "logScale"]:
    
        fig, ax = plt.subplots(nrows=4, ncols=1, sharex=True, sharey=True, figsize=(7, 7))
        
        for ax_i, dataset_name in enumerate(dataset_names):
        
            for pos_i, model_name in enumerate(model_names):
        
                # Get (best) valid. loss
                loader = resultLoader(f"HeRec-{model_name}-{dataset_name}")
                valid_scores = loader.result_df.get_column(f"metrics.{metric}").to_numpy()
                print("Trial #:", len(valid_scores))
                
                # BoxPlot
                ax[ax_i].boxplot( valid_scores, vert=False, widths=0.6, whis=(0, 100), labels=[model_name.replace("_", "-")], positions = [-pos_i] )
        
                # Scatter
                ax[ax_i].scatter( valid_scores, np.ones_like(valid_scores)*(-pos_i), s=20, c="red" )
    
                # Adjust Layout
                if scale_type == "logScale": ax[ax_i].set_xscale( 'log' )
                ax[ax_i].set_xlabel( metric.replace("_", "@") + ' on Valid. Subset' )
                ax[ax_i].set_ylabel( 'Model' )
                ax[ax_i].set_title( dataset_name.replace("_", "-") )
                ax[ax_i].xaxis.set_tick_params(labelbottom=True)
    
        plt.suptitle( scale_type )
        plt.tight_layout()
        plt.savefig( f"01_ValidationLossStability/implicitDataset_boxPlot_{metric}_{scale_type}.pdf" )
        plt.show()
```
